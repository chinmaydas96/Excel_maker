```python
def map_string(value):
    if "AMNET" in value.upper():
        return 'AMNET - DBM'
    
    elif "INSTAGRAM" in value.upper():
        return "Facebook - Instagram"
    
    elif "FACEBOOK" in value.upper():
        return "Facebook"
    
    elif "TABOOLA" in value.upper():
        return "Taboola"
    
    elif "JPN" in value.upper():
        return "DBM_JPN"
    
    elif "VOX MEDIA" in value.upper():
        return "Vox Media"
    
    elif "LINKEDIN" in value.upper():
        return "Linkedin"
    
    elif "CSDN" in value.upper():
        return "CSDN"
    
    else:
        return "Incorrect placement name or new publisher"
    
```

```python
df['Clean Site (DCM)'] = df['Clean Site (DCM)'].apply(map_string)
```
